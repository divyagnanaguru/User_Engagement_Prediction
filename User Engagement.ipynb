```python
import numpy as np
import pandas as pd
```

```python
# Importing the takehome_users.csv into pandas dataframe
data1 = pd.read_csv('takehome_users.csv')
df1 = pd.DataFrame(data1)
df1.head()
```

```python
df1.rename(columns={'object_id':'user_id'}, inplace=True)
```

```python
df1.info()
```

```python
# Converting the datatype of last_session_creation_time from float to datetime
df1['last_session_creation_time'] = pd.to_datetime(df1['last_session_creation_time'])
df1.dtypes
```

```python
df1.describe().T
```

```python
df1.head()
```

```python
# Reading data from takehome_user_engagement.csv
data2 = pd.read_csv('takehome_user_engagement.csv')
data2.head()
```

```python
data2.info()
```

```python
# Converting the datatype of time_stamp from object to datetime
data2.time_stamp = pd.to_datetime(data2.time_stamp)
data2.dtypes
```

```python
new_data = data2.set_index('time_stamp').groupby('user_id')['visited'].resample('D').count()
```

```python
# Importing the formatted data into pandas dataframe
df2 = pd.DataFrame(new_data)
df2.head()
```

```python
# Sum the number of visits for seven-rolling-day period
df2 = df2.rolling(window=7, min_periods=1).sum()
df2.reset_index(inplace = True)
df2.head()
```

```python
# Compute the maximum number of visits for each user for seven-rolling-day period
df3 = df2.groupby('user_id')[['visited']].max()
df3.head()
```

```python
# Adding a new column for adopted user
df3['adopted_user'] = [1 if x >=3 else 0 for x in df3['visited']]
df3.head()
```

```python
# Converting the datatype to int
df3['visited'] = df3['visited'].astype(int)
df3['adopted_user'] = df3['adopted_user'].astype(int)
```

```python
df3.reset_index(inplace=True) 
df3.head()
```

```python
# Merge the two dataframes
final_df = df3.merge(df1, how='inner', on='user_id', sort=True)
```

```python
final_df.head()
```

```python
final_df['user_id'].nunique()
```

```python
final_df1 = final_df[final_df['adopted_user'] == 1]
final_df1.head()
```
